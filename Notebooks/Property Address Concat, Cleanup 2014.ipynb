```python
import pandas as pd
import numpy as np
```

```python
bronx = pd.read_excel('../../Data/Property Data/2014_bronx.xls',header=4)
brooklyn = pd.read_excel('../../Data/Property Data/2014_brooklyn.xls',header=4)
manhattan = pd.read_excel('../../Data/Property Data/2014_manhattan.xls',header=4)
queens = pd.read_excel('../../Data/Property Data/2014_queens.xls',header=4)
statenisland = pd.read_excel('../../Data/Property Data/2014_statenisland.xls',header=4)
```

```python
full=pd.concat([bronx,brooklyn,manhattan,queens,statenisland])
```
