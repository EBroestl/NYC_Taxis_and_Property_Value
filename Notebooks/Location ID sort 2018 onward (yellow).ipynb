```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('../../Data/Taxi Data/yellow_18/yellow_tripdata_2018-12.csv')
```

```python
df
```

```python
df_start=df[['passenger_count','trip_distance','fare_amount','mta_tax','tip_amount','tolls_amount','PULocationID']]
```

```python
df_start=df_start.rename(columns={"PULocationID": "LocationID"})
```

```python
df_end=df[['DOLocationID']]
```

```python
df_start
```

```python
df_end=df.rename(columns={"DOLocationID": "LocationID"})
```

```python
group_start = df_start.groupby(['LocationID']).describe()
```

```python
group_start.columns = ['_'.join(col) for col in group_start.columns.values]
```

```python
group_end = df_end.value_counts(['LocationID'])
```

```python
group_end = pd.DataFrame(group_end,columns=['dropoff_count'])
```

```python
merge=group_start.merge(group_end, how='left', on='LocationID')
```

```python
merge.to_csv('../../Data/yellow_tripdata_2016-07-processed.csv')
```
